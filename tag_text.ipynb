```python
# 提取测试集的标题书名号内文字
all_docs['sep_1']=all_docs['title'].str.extract('[《](.*?)[》]')
all_docs['sep_1']=all_docs['sep_1'].fillna("NoDefine")
word_lib_all_doc_title = all_docs[all_docs['sep_1']!='NoDefine']['sep_1']
word_lib_all_doc_title = word_lib_all_doc_title[word_lib_all_doc_title.str.len()<15].drop_duplicates()
word_lib_all_doc_title.columns = ZuiXinLiuXingXinGe.columns

# 提取测试集的正文书名号内文字
all_docs['sep_2']=all_docs['text'].str.extract('[《](.*?)[》]')
all_docs['sep_2']=all_docs['sep_2'].fillna("NoDefine")
word_lib_all_doc_text = all_docs[all_docs['sep_2']!='NoDefine']['sep_2']
word_lib_all_doc_text = word_lib_all_doc_text[word_lib_all_doc_text.str.len()<15].drop_duplicates()
word_lib_all_doc_text.columns = ZuiXinLiuXingXinGe.columns


# 提取训练集语料库
list_train_docs=[]
for i in range(len(train_docs_keywords['key_words'])):
    list_a = train_docs_keywords['key_words'][i].split(",")
    list_train_docs.extend(list_a)
    
word_train_docs = pd.DataFrame()
word_train_docs['0']=list_train_docs
word_train_docs.columns=ZuiXinLiuXingXinGe.columns

# 合并所有词典
# word_lib=pd.concat([sogou_word_lib,baidu_word_lib,word_lib_all_doc,word_train_docs],axis=0)
word_lib=pd.concat([sogou_word_lib,baidu_word_lib,word_lib_all_doc_title,word_lib_all_doc_text,word_train_docs],axis=0)
# 导出语料库
word_lib=word_lib.reset_index().drop(['index'], axis=1)
word_lib.columns=['word']
word_lib['freq']='3'
word_lib['word']=word_lib['word']+" "+word_lib['freq']
word_lib[['word']].to_csv("./word_library_dic/word_lib.txt",header=None,encoding='utf-8')
```

```python
# 构建分词函数
def tags(text):
    try :
        tags = jieba.analyse.extract_tags(text,topK=2)
#         tags = jieba.analyse.extract_tags(text,topK=5, allowPOS=('ns', 'n', 'vn', 'v'))
#         tags = jieba.analyse.textrank(text, topK=2, allowPOS=('ns', 'n', 'vn', 'v')) 
    except:
        tags = ['','']
    return tags
```

```python
def get_it_1(x):
    try:
        return x[0]
    except:
        return ''    
```

```python
def get_it_2(x):
    try:
        return x[1]
    except:
        return ''    
```
